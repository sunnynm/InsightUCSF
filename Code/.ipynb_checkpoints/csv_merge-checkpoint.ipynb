```python
import pandas as pd
keypd = pd.read_csv('../Data/Cardiolgykeywords.csv', header = None)
keylist = list(keypd.values.flatten())
keylist = [x.lower() for x in keylist]
print(keylist)
for word in keylist:
    for gram in range(1, 4):
        #http://www.albertauyeung.com/post/generating-ngrams-python/
        tokens = [token for token in word.split(" ") if token != ""]
        if len(tokens) < gram:
            break
        ngrams = zip(*[tokens[i:] for i in range(gram)])
        print(["-".join(ngram) for ngram in ngrams])
```

```python
import pandas as pd
all1 = pd.read_csv('../Data/AllAbstracts.csv')

all1["Total"] = (all1["Title"] + all1["Abstract"]).lower()

mllist = ["machine learning","deep learning","artificial intelligence", "ai"]
cardlist = ['cardiology', 'cardiovascular','cardiac','heart']

import csv
counter = 0
with open('../Data/SubAbstracts.csv', mode='w', encoding='UTF-8', newline='') as file:
    ewrite = csv.writer(file, delimiter=',', quotechar='"', quoting=csv.QUOTE_MINIMAL)
    ewrite.writerow(['Source', 'Identifier', 'Title', 'Abstract'])
    
    for index, row in all1.iterrows():
        ml = False
        card = False
        for i in mllist:
            if i in row["Total"]:
                ml = True
        
        for i in cardlist:
            if i in row["Total"]:
                card = True
        if card and ml:
            ewrite.writerow([row["Source"], row["Identifiers"], row["Title"], row["Abstract"]])
            counter += 1
print(counter)
```
